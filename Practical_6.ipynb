```python
# importing numpy  library
# importing pandas library
import numpy as np
import pandas as pd
```

```python
# Loading dataset into a DataFrame
data = pd.read_excel("Sales.xlsx")
data.head(10)
```

##### <span style="color:#003f88">a. Compute mean of a series grouped by another series</span>

```python
grouped_mean = data.groupby('State')['Sales'].mean()

print("The Mean of a Series grouped by another :")
grouped_mean
```

##### <span style="color:#003f88">b. Fill an intermittent time series to replace all missing dates with values of previous non-missing date.</span>

```python
#changing into date format
data['Order Date'] = pd.to_datetime(data['Order Date'])

# Forward-fill missing values
data['Order Date'] = data['Order Date'].ffill()
data.head(10)
```

##### <span style="color:#003f88">c. Perform appropriate year-month string to dates conversion.</span>

```python
#conveting into date time

data['Ship yearMonth'] = pd.to_datetime(data['Ship yearMonth'])
data.head(10)
```

##### <span style="color:#003f88">d.Split a dataset to group by two columns and then sort the aggregated results within the groups.</span>

```python
#lambda function for sorting
func = lambda x: x.sort_values(by='Sales')

sorted_data = data.groupby(['State', 'Region']).apply(func).reset_index(drop=True)
sorted_data.head(10)
```

##### <span style="color:#003f88">e. Split a given dataframe into groups with bin counts.</span>

```python
#creating  bins
bin = pd.cut(data['Sales'], bins=3)

grouped_data = data.groupby(bin, observed=False)
for key, group in grouped_data:
    print("Key : ",key)
    print("GROUP \n",group)
    print("\n\n------------------------------------")
    print("------------------------------------\n\n")
```
