#### <span style="color:#007f5f">4.Consider two excel files having attendance of a workshopâ€™s participants for two days. Each file has three fields , Name ,Time of Joining , duration(in minutes) Where names are unique within a file . Note that duuration may take one of three values (30,40,50) only.Import the data into two dataframes and do the following:</span>

```python
# importing pandas library
# importing numpy library

import pandas as pd
import numpy as np
import openpyxl as op
```

```python
#loading dataset into dataframe
file1='Day1.xlsx'
file2='Day2.xlsx'

df1=pd.read_excel(file1)
df2=pd.read_excel(file2)
```

```python
df1
```

##### <span style="color:#003f88">a. Perform merging of the two dataframes to find the names of students who had attended the workshop on both days</span>

```python
#extracting common attendees
both_day=pd.merge(df1,df2,on='Name',how='inner')
print("Common Attendees : \n")
both_day
```

##### <span style="color:#003f88">b. Find names of all students who have attended workshop on either of the days.</span>


```python
attendees = pd.concat([df1,df2]).drop_duplicates(subset='Name')
print("Total Attendees\n")
attendees
```

##### <span style="color:#003f88">c. Merge two data frames row-wise and find the total number of records in the data frame</span>

```python
merged_df=pd.concat([df1,df2])
print("Total number of records:\n")
len(merged_df)
```

##### <span style="color:#003f88">d. Merge two data frames and use two columns names and duration as multi-row indexes. Generate descriptive statistic for this multi-index.</span>

```python
df3=pd.merge(df1,df2,on=['Name','Duration'],how='inner')
df3

statics=df3.groupby(['Name','Duration']).describe()
print("Descriptive statistic:\n")
statics
```
